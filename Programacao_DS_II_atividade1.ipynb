### Importações

```python
import pandas as pd
 
livros = pd.read_csv('/content/drive/MyDrive/Exercícios /Cópia de livros.csv')

livros_classicos = pd.read_csv('/content/drive/MyDrive/Exercícios /Cópia de livros_classicos.csv')

livros_genero = pd.read_csv('/content/drive/MyDrive/Exercícios /Cópia de livros_genero.csv')
```

### 1) Faça a concatenação das bases de dados e encontre qual o gênero de livro correspondente a cada livro.

```python
result = pd.concat([livros['nome'],livros_genero['genero'],livros_classicos['nome']],axis=1)

display(result)
```

### 2) Faça a concatenação das bases de dados de livros e livros_genero e encontre qual os gêneros e os autores que não possuem um livro correspondente.

```python
livros__genero = pd.concat([livros, livros_genero],axis=1)

mask = result['nome'].isnull()

register = livros__genero[mask]

#print(register)

for index, reg in register.iterrows():
  print(f'O livro de {reg["autor"]} do gênero {reg["genero"]} não possui livro correspondente.')


```

### 3) Faça a junção das bases de livros e livros clássicos.

```python
result = pd.concat([livros, livros_classicos], axis=1)
display(result)
```

### 4) Faça a junção das bases de livros e livros clássicos apenas com os elementos em comum entre ambos e remova os índices dos antigos dataframes.

```python
pd.concat([livros, livros_classicos],axis=0,ignore_index=True,join='inner')

```

### 5) Faça a junção das bases de livros e livros clássicos apenas com os elementos comuns entre ambos, e adicione chaves para o dataframe de livros e de livros clássicos.

```python
pd.concat([livros, livros_classicos],axis=0,keys=['livros','livros_classicos'],join='inner')
```

### 6) Faça a junção das bases de livros e livros clássicos apenas com os elementos comuns entre ambos, e adicione chaves para o dataframe de livros e de livros clássicos, além de nomes para os índices.

```python
pd.concat([livros, livros_classicos],axis=0,keys=['livros','livros_classicos'],join='inner', names=['dataframe','index'])
```
